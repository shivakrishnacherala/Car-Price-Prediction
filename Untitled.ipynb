```python
def get_brand(car_name):
    car_name=car_name.split(" ")[0]
    return car_name.strip()
```

```python
get_brand("shiva krisha")
```

```python
car_data['name']=car_data['name'].apply(get_brand)
```

```python
car_data['mileage']=car_data['mileage'].apply(get_brand)
```

```python
car_data['max_power']=car_data['mileage'].apply(get_brand)
```

```python
car_data['engine']=car_data['engine'].apply(get_brand)
```

```python
for col in car_data.columns:
    print("unique value of "+col)
    print(car_data[col].unique())
    print('-------------------------------\n')
```

```python
unique_names = car_data['name'].unique()
car_data['name'].replace(unique_names, range(len(unique_names)), inplace=True)
```

```python
unique_trans=car_data['transmission'].unique()
car_data['transmission'].replace(unique_trans, range(len(unique_trans)), inplace=True)
```

```python
unique_seller=car_data['seller_type'].unique()
car_data['seller_type'].replace(unique_seller,range(len(unique_seller)),inplace=True)
```

```python
car_data.info()
```

```python
unique_fuel=car_data["fuel"].unique()
car_data['fuel'].replace(unique_fuel,range(len(unique_fuel)),inplace=True)
```

```python
car_data.info()
```

```python
car_data
```

```python
car_data.reset_index(inplace=True)
```

```python
unique_owners = car_data['owner'].unique()
car_data['owner'].replace(unique_owners, range(len(unique_owners)), inplace=True)
```

```python
car_data.drop(columns=['index'],axis=1,inplace=True)
```

```python
corr_matrix=car_data.corr()
print(corr_matrix)
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
threshold = 0.9
high_corr_pairs = [(i, j) for i in corr_matrix.columns for j in corr_matrix.columns if i != j and abs(corr_matrix.loc[i, j]) > threshold]

print("Highly correlated pairs (correlation > 0.9):")
print(high_corr_pairs)
# Remove one of the highly correlated features
# Since max_power and mileage have perfect collinearity (1), we can drop one of them
#cars_data= car_data.drop(columns=['max_power'])
```

```python
input_data=car_data.drop(columns=['selling_price'],axis=1)
output_data=car_data['selling_price']
```

```python
input_data
```

```python
output_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(input_data,output_data,test_size=0.2)
X_train
```

```python
model=RandomForestRegressor(n_estimators=100)
model.fit(X_train,Y_train)
```

```python
X_train.head()
```

```python
y_preds=model.predict(X_test)
y_preds
```

```python
model.score(X_test,y_preds)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(Y_test,y_preds)
r2
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(X_train,Y_train)
```

```python
y_preds2=model1.predict(X_test)
y_preds2
```

```python
r21=r2_score(Y_test,y_preds2)
r21
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model3=DecisionTreeRegressor()
model3.fit(X_train,Y_train)
```

```python
y_preds3=model3.predict(X_test)
y_preds3
```

```python
model3.score(X_test,y_preds3)
```

```python
r23=r2_score(Y_test,y_preds3)
r23
```
